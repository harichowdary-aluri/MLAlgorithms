```python
#converting str data type to Timestamp
df=pd.read_csv("weather.csv",parse_dates=['Date'])
type(df.Date[0])
```

```python
df.set_index('Date',inplace=True)
df
```

```python
newdf=df.fillna(0)
newdf
```

```python
#we can use dictionary to fill specific value to a specific columns
newdf = df.fillna({
    'Evaporation':0,
    'Sunshine':'no sunshine'
})
newdf
df
```

```python
#it fill with previos value when it encounter NAN
newdf=df.ffill()
# we can set the limit if you want
newdf=df.ffill(limit=1)
newdf
```

```python
newdf=df.interpolate()
newdf
#or
newdf=df.infer_objects(copy=False)
newdf
```

```python
#TO DROP NA VALUES 
newdf=df.dropna()
newdf
```

```python
#TO DROP NA VALUES (IF ALL THE rows ARE empty)
newdf=df.dropna(how="all")
newdf=df.dropna(thresh=2)
newdf
```

```python
#if i want to have all the dates in a data set but the dates are missing in it 
dt=pd.date_range("2013-03-13","2008-12-04")
idx=pd.DatetimeIndex(dt)
df=df.reindex(idx)
df
```
