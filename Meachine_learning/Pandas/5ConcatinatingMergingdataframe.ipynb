```python
import pandas as pd

indian = pd.DataFrame({
    'city': ['mumbai', 'delhi', 'bangalore'],
    'temperature': [20, 30, 40],
    'humidity': [80, 60, 78]
})

indian1 = pd.DataFrame({
    'city': ['mumbai', 'delhi', 'bangalore'],
    'windspeed': [34444, 3443550, 423432420]
})

# Concatenate DataFrames along axis 0 (rows)
df = pd.concat([indian, indian1], axis=1)   # axis =0 kuda pettu

# Print the DataFrame
print(df)
```

```python
# Merge uses intersection if the data in one df present in the another df than only it merges or else it will leave
indian = pd.DataFrame({
    'city': ['mumbai', 'delhi', 'bangalore'],
    'temperature': [20, 30, 40],
    'humidity': [80, 60, 78]
})

indian1 = pd.DataFrame({
    'city': ['mumbai', 'bangalore', 'delhi'],
    'windspeed': [34444, 3443550, 423432420]
})
df=pd.merge(indian,indian1,on='city')
df
```
