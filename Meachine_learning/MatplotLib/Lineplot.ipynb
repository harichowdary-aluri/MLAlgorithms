```python
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
print(plt.plot(batsman['index'],batsman['V Kohli']))
```

```python
print(plt.plot(batsman['index'],batsman['RG Sharma']))
```

```python
#plotting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
#labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohith vs Virat career')
plt.xlabel('IPL seasons')
plt.ylabel('Runs Scored')
```

```python
#coloring
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')
plt.title('Rohith vs Virat career')
plt.xlabel('IPL seasons')
plt.ylabel('Runs Scored')
```

```python
#line styling
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dotted',linewidth=3,marker='D',markersize=10)#dashdot #d = dimond o=circle
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed',linewidth=2,marker='o')
plt.title('Rohith vs Virat career')
plt.xlabel('IPL seasons')
plt.ylabel('Runs Scored')
```

```python
#legend->location
#e line evaru ani theliyadaniki prati plot lo label use chestham and a labels display avvadanki legend() ni call chestham 
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dotted',linewidth=3,marker='D',markersize=10,label='kohli')#dashdot #d = dimond o=circle
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed',linewidth=2,marker='o',label='Sharma')
plt.title('Rohith vs Virat career')
plt.xlabel('IPL seasons')
plt.ylabel('Runs Scored')
plt.legend(loc='best')#best antey automactic ga kali vunna chota kurchuntundhi 
loc= upper right upper left lower right lower left
#right center left center right lower center
```
