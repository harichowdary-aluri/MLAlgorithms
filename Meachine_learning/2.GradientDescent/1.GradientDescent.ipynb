```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

n_samples: This parameter specifies the number of samples (data points) you want to generate. In this case, you're generating a dataset with 4 samples.

n_features: It specifies the number of features (independent variables) in your dataset. Here, you're generating a dataset with 1 feature.

n_informative: This parameter determines the number of informative features, i.e., the number of features that are actually used to generate the output. In your case, 1 feature is informative, meaning it is relevant to the target variable y.

n_targets: It specifies the number of target variables (dependent variables) you want to generate. Here, you're generating 1 target variable.

noise: This parameter controls the amount of noise added to the output. A higher value indicates more noise. In your case, the noise level is set to 80.

random_state: This parameter is used to control the randomness of the data generation process. By setting random_state to a specific value (in this case, 13), you ensure that every time you run the code, you'll always get the same dataset. It helps in making the results reproducible.

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# lets apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')
```

```python
# Lets apply Gradient descent assuming slope is constant m=78.35
# let's assume the starting value for intercept b=0
#ippudu manam just oka normal line giddam which is b=0
#y=mx=c
ypred=78.35*x+0
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,ypred,color='blue',label='b=0')
plt.legend()
```

```python
#Original array: [[1 2 3] [4 5 6]]   Flattened array: [1 2 3 4 5 6]
m=78.35
b=0
loss_slop=-2*np.sum(y-m*x.ravel()-b)
loss_slop
```

```python
# Lets take learning rate as 0.1
lr=0.1
step_size=loss_slop*lr
b = b-step_size
b
```

```python
ypred1=78.35*x+b
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,ypred1,color='black',label='b={}'.format(b))
plt.legend()
plt.show()
```
